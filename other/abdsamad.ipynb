```python
import pandas as pd
import numpy as np
```

```python
#read the dataset
boston_dataset = pd.read_csv('boston.csv')
```

```python
#display the first 5 rows of the dataset
boston_dataset.head()
```

```python
#display dataset information
print(boston_dataset.info())
```

```python
#delete all collumns with url or id in the name
for key, value in boston_dataset.items():
    if "url" in key or "id" in key:
        del boston_dataset[key]
print("number of columns : ", boston_dataset.shape[1])
```

```python
#Count nulls percentages in each column
nulls_percentages = boston_dataset.isnull().sum()/len(boston_dataset)*100
#display the nulls percentages
for key, value in nulls_percentages.items():
    print(key,"\t<====>\t", value)
```

```python
#delete all columns with more than 30% of nulls
for key, value in nulls_percentages.items():
    if value > 30:
        del boston_dataset[key]

print("number of columns : ", boston_dataset.shape[1])
```

```python
#show for every variable its unique values
for key, value in boston_dataset.items():
    print("==================",key, "==================\n", value.unique())
    print("--------------------------------------------\n")
```

```python
print("number of columns before : ", boston_dataset.shape[1])
del boston_dataset['last_scraped'] #not useful for the analysis (only one value)
del boston_dataset['host_name'] #not useful for the analysis (nobody cares about the host name)
del boston_dataset['zipcode'] #not useful for the analysis
del boston_dataset['description'] #not useful for the analysis (too much text)
del boston_dataset['summary'] #not useful for the analysis (too much text)
del boston_dataset["house_rules"] #not useful for the analysis (too much text)
del boston_dataset["name"] #not useful for the analysis (too much text)
del boston_dataset["smart_location"] #not useful for the analysis (duplicate of city)
del boston_dataset["street"] #not useful for the analysis (duplicate of city)
del boston_dataset["neighborhood_overview"] #not useful for the analysis (too much text)
del boston_dataset["experiences_offered"] #not useful for the analysis (only one value (None))
del boston_dataset["host_location"] #not useful for the analysis (too much text)
del boston_dataset["space"]    #not useful for the analysis (too much text)
del boston_dataset["calendar_last_scraped"] #not useful for the analysis (only one value)
del boston_dataset["host_neighbourhood"] #not useful for the analysis (too much text)
del boston_dataset["host_total_listings_count"] #not useful for the analysis (duplicate of host_listings_count)
del boston_dataset["latitude"] #not useful for the analysis 
del boston_dataset["longitude"] #not useful for the analysis
del boston_dataset["calendar_updated"]
print("number of columns after : ", boston_dataset.shape[1])
```

```python
#delete all columns with only one value
print("number of columns before : ", boston_dataset.shape[1])
for key, value in boston_dataset.items():
    if len(value.unique()) == 1:
        del boston_dataset[key]
print("number of columns after : ", boston_dataset.shape[1])
```

```python
#show the first to values of all variables that has review in the name
for key, value in boston_dataset.items():
    if "review" in key:
        print(key, "==================\n", value.head(2))
        print("-----------------\n")
```
