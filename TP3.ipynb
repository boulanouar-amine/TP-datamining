K-means Clustering </center> 


```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

- lecture de l'image

```python
image = Image.open('data/TP3/bird_small.png')
image = np.array(image)
```

-  Normalisation des données

```python
image = image / 255
```

- creation d'une liste de pixel

```python
pixels = image.reshape(-1, 3)
```

- appliquation du KMeans

```python
kmeans = KMeans(n_clusters=16)
kmeans.fit(pixels)
```

- on remplace chaque pixel par la moyenne du centroïde 

```python
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = compressed_pixels.reshape(image.shape)
```

- affichage de l'image original et compressee

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6),
                       subplot_kw=dict(xticks=[],
                                       yticks=[],
                                       frame_on=False))
ax[0].imshow(image)
ax[0].set_title('Image Original')
ax[1].imshow(compressed_image)
ax[1].set_title('Image Compressee')
plt.show()
```
