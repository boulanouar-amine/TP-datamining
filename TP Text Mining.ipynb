# Import Required Libraries
Import the necessary libraries, including pandas, nltk, etc.

```python
# Importing the necessary libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Downloading the necessary nltk packages
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

# Load Comments Dataset
Load the comments dataset using pandas.

```python
# Load Comments Dataset
comments_original = pd.read_csv('NXpdyAWLDas_comments.csv')

# Display the first few rows of the dataset
comments_original.head()
```

# Text Lowercasing
Convert all the text to lower case. This is required as python interprets 'dog' and 'DOG' differently.

```python
# Convert all the text to lower case
comments_df = pd.DataFrame()
comments_df['comment'] = comments_original['comment'].str.lower()



# Display the first few rows of the dataset

comments_df.head()
```

# Removing Punctuation
Remove all punctuation from the text as it doesn’t add any extra information while treating text data.

```python
# Importing the string library
import string

# Function to remove punctuation


def remove_punctuation(text):
    # Using the translate() method to remove punctuation
    return text.translate(str.maketrans('', '', string.punctuation))


# Applying the function to the 'comment' column
comments_df['comment'] = comments_df['comment'].apply(remove_punctuation)

# Display the first few rows of the dataset
comments_df.head()
```

# Removing Stop Words
Remove commonly used words (such as 'is', 'the', 'a') as they will likely appear in any text.

```python
# Define the stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stop_words(comment):
    tokens = word_tokenize(comment)
    filtered_comment = [word for word in tokens if word not in stop_words]
    return ' '.join(filtered_comment)

# Apply the function to the 'comment' column
comments_df['comment'] = comments_df['comment'].apply(remove_stop_words)

# Display the first few rows of the dataset
comments_df.head()
```

# Stemming
we are going to skip stemming and just use lemmatization

# Lemmatization
Group together the inflected forms of a word so they can be analysed as a single item. This is similar to stemming but it brings context to the words.

```python
# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to apply lemmatization to a list of words
def lemmatize_words(text):
    words = word_tokenize(text)
    return ' '.join([lemmatizer.lemmatize(word) for word in words])

# Apply the function to the 'comment' column
comments_df['comment'] = comments_df['comment'].apply(lemmatize_words)

# Display the first few rows of the dataset
comments_df.head()
```

```python
# Importing the PorterStemmer from nltk
from nltk.stem import PorterStemmer

# Creating an object of class PorterStemmer
porter = PorterStemmer()

# Defining a function to perform stemming on the comments
def stem_words(text):
    words = word_tokenize(text)
    stemmed_words = [porter.stem(word) for word in words]
    return " ".join(stemmed_words)

# Applying the stemming function to the comments
comments_df['comment'] = comments_df['comment'].apply(stem_words)

# Display the first few rows of the dataset
comments_df.head()

# save 
comments_df.to_csv("cleaned_comments.csv", index=False)
```

```python
# show an example of the original comment and the cleaned comment
print('Original Comment:', comments_original['comment'][4])
print('Cleaned Comment:', comments_df['comment'][4])
```

# Tokenization

```python
# Assume df['comment'] is the column with the text you want to tokenize
comments_df['tokenized_comment'] = comments_df['comment'].apply(
    nltk.word_tokenize)

# group all the comments into one list of words
words_list = comments_df['tokenized_comment'].values
words_list = [word for comment in words_list for word in comment]
print(len(words_list))
```

```python
words_list = set(words_list)
print(len(words_list))
```

# Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create an instance of CountVectorizer
vectorizer = CountVectorizer(vocabulary=words_list)

# Fit the vectorizer to the comments and transform the comments into a 'bag of words'
X = vectorizer.fit_transform(comments_df['comment'])

# Convert the matrix to an array and print its shape
X = X.toarray()
print(X.shape)
```

# Plot the word frequencies

```python
# plot the word frequency
import matplotlib.pyplot as plt

# Calculate the frequency of each word
word_frequencies = X.sum(axis=0)

# Create a dictionary with the words as keys and the frequencies as values
word_freq_dict = dict(zip(words_list, word_frequencies))

# Sort the dictionary by the frequencies in descending order
sorted_word_freq_dict = dict(
    sorted(word_freq_dict.items(), key=lambda x: x[1], reverse=True))

# Create a list of the top 10 most frequent words
top_10_words = list(sorted_word_freq_dict.items())[:10]

# Plot the top 10 most frequent words
plt.figure(figsize=(10, 5))

# Create a bar plot
plt.bar([word[0] for word in top_10_words], [word[1] for word in top_10_words])

# Add labels and title
plt.xlabel('Words')
plt.ylabel('Frequency')

plt.title('Top 10 Most Frequent Words in the Comments')

# Display the plot
plt.show()
```

```python
from nltk import bigrams, FreqDist

# Tokenize the text into words
words = [word for comment in comments_df['comment'] for word in nltk.word_tokenize(comment)]

# Generate bigrams
bigram_list = list(bigrams(words))

# Create a frequency distribution of the bigrams
bigram_freq = FreqDist(bigram_list)

bigram_freq
```

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

def get_most_common_bigram_stemmed(text):
    # Stem the input text
    stemmed_text = ' '.join([stemmer.stem(word) for word in nltk.word_tokenize(text)])
    
    # Get the last word of the stemmed text
    last_word = stemmed_text.split()[-1]
    
    # Find all bigrams that start with the last word
    bigrams_starting_with_last_word = [bigram for bigram in bigram_list if stemmer.stem(bigram[0]) == last_word]

    # Create a frequency distribution of these bigrams
    bigram_freq = FreqDist(bigrams_starting_with_last_word)

    # Return the three most common bigrams
    return bigram_freq.most_common(3)

# Test the function
print(get_most_common_bigram_stemmed('but I reckon in another'))
```
