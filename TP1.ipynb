## Gathering the columns that we will drop

```python
# create an array to store the columns that shoud be deleted
columns_to_delete = []
```

- columns with url

```python
# columns with url in the name
columns_with_url = [col for col in df.columns if 'url' in col]
# print columns not in columns_to_delete
print([col for col in columns_with_url if col not in columns_to_delete])
columns_to_delete.extend(columns_with_url)
```

- columns with dates

```python
date_columns = ['last_scraped', 'host_since', 'calendar_last_scraped', 'first_review', 'last_review']
```

- columns with textual data

```python
# get all columns with string data type 
text_cols = df[df.columns[df.dtypes == 'object']] 
is_text = []

for col in text_cols:
    # if it is inuque 70% of the time
    if df[col].nunique() / df.shape[0] > 0.1:
        is_text.append(col)
print(len(is_text))
print([col for col in is_text if col not in columns_to_delete])

# don't delete the date_columns
is_text = [col for col in is_text if col not in date_columns]

columns_to_delete.extend(is_text)
```

- columns that can be calculated

```python
# columns that can be calculated from other columns
calculated_cols = ['maximum_nights_avg_ntm', 'minimum_nights_avg_ntm', 'maximum_maximum_nights',
                   'minimum_maximum_nights', 'maximum_minimum_nights', 'minimum_minimum_nights',
                   'availability_60', 'availability_90', 'availability_365']

print([col for col in calculated_cols if col not in columns_to_delete])
columns_to_delete.extend(calculated_cols)
```

- columns with id 

```python
# columns with id in the name
# ignore host_identity_verified
columns_with_id = [col for col in df.columns if 'id' in col and col != 'host_identity_verified']
print(columns_with_id)
columns_to_delete.extend(columns_with_id)
```

- columns duplicated of another column

```python

# duplicate columns of other columns

duplicate_columns = []

for col in df.columns:
    for col2 in df.columns:
        if col != col2 and df[col].equals(df[col2]) and col not in duplicate_columns:
            duplicate_columns.append(col)
print(len(duplicate_columns))
print([col for col in duplicate_columns if col not in columns_to_delete])

columns_to_delete.extend(duplicate_columns)
```

- cloumns with null values

```python
# columns with all null values or same value on all rows
null_columns = df.columns[df.isnull().all()]

print([col for col in null_columns if col not in columns_to_delete])
columns_to_delete.extend(null_columns)
```

- columns with same value in all rows

```python
# columns with same value on all rows
same_value_columns = df.columns[(df.nunique() == 1)]
print([col for col in same_value_columns if col not in columns_to_delete])
columns_to_delete.extend(same_value_columns)
```

- columns with less than 50% of values

```python

# columns with more than 50% of the values missing
missing_values = df.columns[(df.isnull().sum() / df.shape[0]) > 0.5]
print([col for col in missing_values if col not in columns_to_delete])
columns_to_delete.extend(missing_values)
```

- irrelevant data


```python
# irrelevant columns
irrelevant_columns = ['zipcode', 'requires_license',
                      'host_verifications',"street", 'calendar_updated', "neighbourhood_cleansed","city", "smart_location"]
print([col for col in irrelevant_columns if col not in columns_to_delete])
columns_to_delete.extend(irrelevant_columns)
```

- columns with over 90% of repeated values

```python
# columns with over 90% of the values the same

same_value_columns =[]

for col in df.columns:
    # calculate the percentage of the most common value
    values = df[col].value_counts()
    # check if it is over 90%
    for value in values:
        if value / df.shape[0] > 0.9:
            same_value_columns.append(col)
    
print([col for col in same_value_columns if col not in columns_to_delete])

columns_to_delete.extend(same_value_columns)
```

- `city` before deleting this column, we will keep only the rows from boston.  
`host_location` we only consider the listings in Boston and Cambridge, so this column is irrelevent.

```python
df = df[df["city"].fillna("").str.lower().str.contains("boston") | df["city"].fillna("").str.lower().str.contains("cambridge")]
columns_to_delete.append("host_location")
```

* columns that appear to be invalide but they are not, so we will keep them:
    - latitude

    - amenities



```python
columns_to_delete = [col for col in columns_to_delete if col not in ['latitude', 'longitude', 'amenities']]
```

```python
print("old shape: ", df.shape)
df = df.drop(columns=columns_to_delete)
print("new shape: ", df.shape)
```

## Handeling date values

```python
# all the date columns that are left
for col in [col for col in date_columns if col not in columns_to_delete]:
    df[col] = pd.to_datetime(df[col])
```

## Handling missing values

```python
missing = df.isna().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_percentage = (missing / df.shape[0]) * 100
missing = pd.DataFrame(
    {'missing': missing, 'Percentage %': round(missing_percentage, 2)})
missing
```

filling the columns that have few missing values with the average

```python
for col in missing.index:

    if missing.loc[col, "Percentage %"] < 5:

        # if it's a str column, fill with the most common value
        if df[col].dtype == "object":
            value = df[col].mode()[0]
            print(
                f"filling {col} with the most common value {value}")
            df[col] = df[col].fillna(df[col].mode()[0])

        # if it's a numerical column, fill with the most common value

        if df[col].dtype == "int" or df[col].dtype == "float":
            value = round(df[col].mean())
            print(
                f"filling {col} with the mean value {value}")

            df[col] = df[col].fillna(df[col].mean())
```

```python
# fill the rest with unknown
df["host_response_time"] = df["host_response_time"].fillna("unknown")

# assume that the missing values mean that the host does not have a review yet or a security deposit etc...
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
df["security_deposit"] = df["security_deposit"].fillna(0)
df["cleaning_fee"] = df["cleaning_fee"].fillna(0)

# drop the rows without the host since
df = df.dropna(subset=["host_since"])

# update the missing values
missing = df.isna().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_percentage = (missing / df.shape[0]) * 100
missing = pd.DataFrame(
    {'missing': missing, 'Percentage %': round(missing_percentage, 2)})
missing
```

## Fixing boolean columns

```python
# if the column contains "t","f" we can convert it to bool
bool_cols = []
for col in df.columns:
    if df[col].nunique() == 2 and "t" in df[col].unique() and "f" in df[col].unique():
        bool_cols.append(col)

bool_cols
```

```python
# convert the columns to bool
for col in bool_cols:
    df[col] = df[col].map({"t": True, "f": False})
    df[col] = df[col].astype("bool")
```

## Fixing currency cols

```python
# currency cols are the one that have rows that start with $
currency_cols = []
for col in df.columns:
    if df[col].dtype == "object" and df[col].str.startswith("$").any():
        currency_cols.append(col)
        
currency_cols
```

```python
# remove the $ and convert to float
for col in currency_cols:
    df[col] = df[col].str.replace("$", "").str.replace(",", "").astype(float)
    # fill the missing values with 0
    df[col] = df[col].fillna(0)
```

## Other fixes

```python
# having 1.2 bathrooms doesn't make sense
df["bathrooms"] = df["bathrooms"].astype(int)

# fixing percentage in host_response_rate to a float
df["host_response_rate"] = df["host_response_rate"].str.replace("%", "").astype(float)
df["host_response_rate"] = df["host_response_rate"].fillna(0)
```

## Handeling amenities

```python
import re
from itertools import chain

# Use regular expression to remove unwanted characters
df["amenities"] = df["amenities"].str.split(",")

# remove {,}," and extra spaces
df["amenities"] = df["amenities"].apply(
    lambda x: [re.sub(r'["{}]', "", a).strip() for a in x])

# remove empty strings and translation missing
df["amenities"] = df["amenities"].apply(
    lambda x: [a for a in x if a and not a.startswith("translation missing: ")])

# flatten the list of lists
df_amenities = pd.Series(list(chain.from_iterable(
    df["amenities"].tolist()))).value_counts()
df_amenities = df_amenities.reset_index()  # convert to dataframe
df_amenities.columns = ["amenities", "count"]  # rename columns

df_amenities
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# remove the amenities that less than 10% from the number of the rows
amenities_threshold_min = df.shape[0] * 0.1
amenities_threshold_max = df.shape[0] * 0.9
print(amenities_threshold_min, amenities_threshold_max, df.shape[0])
insufficient_amenities = df_amenities[df_amenities["count"] < amenities_threshold_min]
repeatitive_amenities = df_amenities[df_amenities["count"] > amenities_threshold_max]


plt.figure(figsize=(10, 10))
sns.barplot(data=df_amenities, y="count", x="amenities")
plt.title("Amenities in Boston AirBnB")
plt.xticks([])
plt.axhline(y=amenities_threshold_min, color='r', linestyle='--')  # Add minimum threshold line
plt.axhline(y=amenities_threshold_max, color='g', linestyle='--')  # Add maximum threshold line

# Add threshold labels at the rightmost positions
plt.text(plt.xlim()[1], amenities_threshold_min, 'Amenities Threshold (Min)', color='r', va='bottom', ha='right')
plt.text(plt.xlim()[1], amenities_threshold_max, 'Amenities Threshold (Max)', color='g', va='bottom', ha='right')

plt.show()
```

```python
# remove them from the dataframe
# insufficient amenities
df_amenities = df_amenities[~df_amenities["amenities"].isin(
    insufficient_amenities["amenities"])]
# repeatitive amenities
df_amenities = df_amenities[~df_amenities["amenities"].isin(
    repeatitive_amenities["amenities"])]

print("columns count : ", df_amenities.shape[0])
df_amenities
```

```python
# one hot encoding based on df_amenities
df_amenities = df_amenities["amenities"].tolist()

df = df.join(df["amenities"].apply(lambda x: pd.Series(
    [True if a in x else False for a in df_amenities], index=df_amenities)))

df.drop(columns=["amenities"], inplace=True)

df.shape
```
